On the top level data can be categorised in two types :-


   __Structure Data__  
   __Unstructure Data__ 

### Structure Data

which follow a pattern and have well define structure . like spreadsheet , CSV file , SQL files etc. In general this type of data structure contain the following types of data values (variable).

```python
display(Image(filename = 'Data Type.png'))
```

Quantitative - Which describe the quantity of the variable like No. of House , Marks in particular Subject.

Now the Quantitative type can further classified in two groups .

a) Discrete - where the value is in integer , can't be express in fractional like No. of House , No. of Children etc.

b) Continuous - Where the value can be both integer or fractional . like Marks in an exam , Height of a person , weight of a person .

Qualitative - A non numeric value which describe the quality of a variable like Gender , marital status , Educational Qualification etc.

Qualitative type of data also classified in two groups.

a) Nominal - where ranking or hierarchy is not possible like male or female, if for calculation we put male = 1 & female = 2 it doesn't mean that female is superior .

b) Ordinal - Even describing the quality it follows a hierarchy , like a coloum of Educational Qualification ,  graduate, post graduate , PHD . we don't have number but a PHD is always bigger then graduate.



With structure dataset and with these variable type we can performe __Supervised & Unsupervised__ Learning

### Unstructure Data

A dataset which contain lots of information but doesn't follow any pattern or structure. like facebook post , tweets , Photographs , Voice recordings , Songs etc.

To get information from Unstructure Data we use __neural network & Deep learning__  Algorithm.

As we have to perform the task in computer with the help of Programing Language so before we start its better to understand the Data Structure of a programing Language

__Built-in Data Structure__ are basicly the variable types that a computer supports .

- Integer - Number without fraction or dicimal value . i.e - 2,3,5,7
- Float - Number with dicimal value or fraction . i.e - 3.576 , 9.123, 7.54
- Character - Alphabets. i.e chandan , jhon , M.B.A
- Boolean - True / False , Yes/No 
- Pointer - References a location in memory(not datatype)

__User Define Data Structures__ are database type which we are going to deal with Machine Learning algrothims. 

- __Array__ - Arrays are fixed size linear data structures in which the elements are accessed by indicating its index value. The index value of the starting element of an array will be 0.

- __Lists__ - It is same as array but of any shape and size , list can be further classified into many parts.

- __Files__ - when data is store in pre define format , with diferent delimeters . i.e Spreadsheets , CSV, Flat files etc.
