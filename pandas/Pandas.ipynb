- As we see are in object format so first we have to map them then we can go for further analysis

```python
# mapping gender 
def map_gender(gender):
    if gender == 'M':
        return 0
    else:
        return 1
db['gender'] = db['gender'].apply(map_gender)
```

```python
# mapping age 
def map_age(age):
    if age == '0-17':
        return 0
    elif age == '18-25':
        return 1
    elif age == '26-35':
        return 2
    elif age == '36-45':
        return 3
    elif age == '46-50':
        return 4
    elif age == '51-55':
        return 5
    else:
        return 6
db['age'] = db['age'].apply(map_age)
```

```python
# mapping city Category
def map_city_categories(city_category):
    if city_category == 'A':
        return 2
    elif city_category == 'B':
        return 1
    else:
        return 0
db['city_category'] = db['city_category'].apply(map_city_categories)

```

```python
# mapping current city (4+ is creating issue)
def map_stay(stay):
        if stay == '4+':
            return 4
        else:
            return int(stay)

db['current_city'] = db['current_city'].apply(map_stay)  
```
