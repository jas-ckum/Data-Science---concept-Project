100 % accuracy is myth .

Statistics does not study individuals. 


__Law of Statistical Regularity__

1 ) The sample should be chosen from the population at random. That is , every individual of the population should have equal chance of inclusion in the sample .

2 ) The size of the sample should be moderately large . The larger the sample size , the greater is the accuracy .

 __Important terms in Statistics__.

- Hypothesis : A subjective statment based on which we start building our model.

- Data : The collection of numerical figures .

- Population : This is the totality of the setof object or individuals under consideration .

- Sample : A sample is a selected number of individuals each of which is chosen from the corresponding population.

- Characteristics : This is the quality possessed by an individual item of population.

- Variable : This is the characteristics of an individual which can be measured numerically in terms of units .

- Attributes : This is the characteristics of an individual which cannot be measured numerically .

- Continuous Variable : This is the variable which can assume all possible value within a specified range .

- Discrete Variable : This is the variable which can assume only some specified isolated values .

- Dependent Variable : The variable which we are trying to measure or predict.

- Independent Variable : The variable by which we can measure dependent variable .

- Bias : opportunity cost or alternative cost. 


** __IN Stats we only deal with two types of variable__. i.e 

Continuous Variable

Discrete Variable

__Statistics data can classified in Two types__.

- Primary Data : Data which are collected for first time , for some specific purpose & directly from the field of enquiry .

- Secondary Data : Data which are obtained from published or Unpublished sources . 

** In data Science we will mostly dealing with Secondary data. 

__Error in Secondary Data__

- Transcribing error : Every time figure is copied , there is a chance of error.

- Estimating error : Sometimes , the figure are put in the secondary data by mere estimation. This is sometime needed because the purpose of collection of primary data is not same as that of the secondary data.

- Error due to bias : Sometimes the statistician may be biased and may put some fictitious figure in the secondary data .

### Descriptive analysis

- The process in which we describe the variable . Descriptive analysis further classified into three parts 

__Univariate analysis__

__Bivariate analysis__

__Multivariate analysis__

#### Univariate analysis

- Is the simplest from of data analysis .where the data being analyzed contains only one variable . The statistical tools for univariate analysis are :-

a) Mean 

b) Median 

c) Mode

d) Range 

e) Variance 

f) Maximum

g) Minimum 

h) quartiles

i) Standard deviation

- The graphs that we use for univariate analysis are :

a) Bar Charts

b) Histograms

c) frequency distribution

d) Pie charts.

####  Bivariate analysis

- It is use to find out the relationship between two different variable. mainly we try to established the relationship between Dependent variable & Independent Variable. The statistical tools for Bivariate analysis are :

a) Chi - Square 

b) Anova 

c) correlation

d) T - test. etc


These test are based on the variable types for Dependent variable & Independent Variable, to understand this have a look on below graph.

__Chi-Square__ : chi-squared test is used to determine whether there is a significant difference between the expected frequencies and the observed frequencies in one or more categories .

__ANOVA__  : (Analysis of variance ) is determine the "variation" among and between the groups .

__Correlation__ : It measure the association between two continuous variable . Its value range is between -1 to +1 . 

__T-Test__ : It measure the significant difference between the means of two groups . it is use when our sample size is less the 30 , for sample size more then 30 we use __Z-test__ .

####  Multivariate analysis

- It is use to find out the relationship between three or more variable . The statistical tools for Multivariate analysis are :

a) Additive tree 

b) cluster analysis

c) Manova 

d) Multidimensional scaling .etc


- usage of Multivariate analysis is very rare .

### Predictive Analysis

- Once we find out the main independent variable in the dataset using   Bivariate analysis & Multivariate analysis , now its time to predict the dependent variable using the tools of Predictive Analysis .

- Predictive Analysis further classified into three parts .


__Clustering__ : Its a process to grouping all the object of same set .

__Classification__ : when our dependent variable is Discrete we use classificaton model.

__Regression__ : : when our dependent variable is continuous we use Regression model .
